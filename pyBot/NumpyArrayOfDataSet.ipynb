```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from numpy import asarray


# Get labels form csv file
train_csv = "drive/MyDrive/AI/train_no_1.csv" #train_all.csv and train_no_0.csv
df_train = pd.read_csv(train_csv)

# Numpy array will contain images
train_X = []

# DataFrame will contain labels
df_train_Y = []

# Array that will be coverted to numpy array or dataFrame
train_X_array = []
train_Y_array = []

# Fixed values
IMG_SIZE = 200
```

```python
# We loop through the photo's and transform them to a numpy array

# Set train labels and images into an array
def set_training_data():
  # Loop the train Dataframe
  for index, row in df_train.iterrows():
    try:
      # Open, resize and convert image to array
      img_train = Image.open(row.path)
      plt.imshow(img_train)
      img_train = img_train.resize((IMG_SIZE, IMG_SIZE))
      np_img_train = asarray(img_train)

      # Put values in arrays
      train_X_array.append(np_img_train)
      train_Y_array.append([row.ethnicity,row.gender,row.age])

    except Exception:
      pass

set_training_data();
```

```python
# After the loop is completed the array needs to be a numpy array and DataFrame
train_X = np.array(train_X_array)
train_X.shape

#The X values are saved as a numpy array
temp_X = np.array(train_X_array)
np.save("np_train_X_no_1", temp_X)

# The Y values are saved as .CSV file
temp_Y = pd.DataFrame(train_Y_array, columns=["ethnicity","gender","age"])
temp_Y.to_csv('df_train_Y_no_1.csv', index = True)
```
