```python
# In this file the dataset.csv will be split into test and train set. The train set will be saved  3 times with all the ethnicity's, without ethnicity white and without ethnicity black.
# Google Collab was used for this project. The used photo's and files are not included in this project because of the size.
```

```python
#import libraries
import pandas as pd

#specify the path
path_img = "drive/MyDrive/UTKFace/"
path_csv = "drive/MyDrive/data_tot.csv"

#read labels
df_labels = pd.read_csv(path_csv)
```

```python
# We only will use photos of people with ethnicity 0,1 or 2
df_used = pd.concat([df_labels.loc[df_labels["ethnicity"] == 2], df_labels.loc[df_labels["ethnicity"] == 1], df_labels.loc[df_labels["ethnicity"] == 0]])
print(df_used.shape)
```

```python
# The shape of df_used is 17918 This divided by 4 gives 4479,5
# The set will be shuffled to make the test and train set as equal as possible
df_used = df_used.sample(frac=1).reset_index(drop=True)
# 1/4 will be saved as test set
df_test = df_used[:4480]
# 3/4 will be saved as train set
df_train = df_used[4480:]

df_test.to_csv('test.csv', index = True)
df_train.to_csv('train.csv', index = True)
```

```python
# training set will be loaded from the .CSV file
train_csv = "drive/MyDrive/AI/train.csv"

# .CSV is converted into a DataFrame
df_train = pd.read_csv(train_csv)

# not used columns are deleted
del df_train['Unnamed: 0']
del df_train['Unnamed: 0.1']

# The 3 different train sets are made and saved
df_train_no_ethnicity_0 = pd.concat([df_train.loc[df_train["ethnicity"] == 2], df_train.loc[df_train["ethnicity"] == 1]])
df_train_no_ethnicity_1 = pd.concat([df_train.loc[df_train["ethnicity"] == 2], df_train.loc[df_train["ethnicity"] == 0]])
df_train.to_csv('train_all.csv')
df_train_no_ethnicity_0.to_csv('train_no_0.csv')
df_train_no_ethnicity_1.to_csv('train_no_1.csv')
```
