Photo's are get from https://susanqq.github.io/UTKFace/ and the CSV file is from https://www.kaggle.com/nipunarora8/age-gender-and-ethnicity-face-data-csv

```python
# import libraries
import glob
import os
import pandas as pd

#specify the path
path_img = "drive/MyDrive/UTKFace/"
path_csv = "drive/MyDrive/age_gender.csv"

#read labels
labels = pd.read_csv(path_csv)
image_labels = pd.concat

#get the filenames
jpgFilenamesList = glob.glob(f"{path_img}/*")
```

```python
# the given path of the images in .CSV file will be altered
combine = []

for index, row in labels.iterrows():
    for path in jpgFilenamesList:
      if os.path.basename(path).split("_")[-1] == row.img_name:
        combine.append([path,row.ethnicity,row.gender,row.age])
        jpgFilenamesList.remove(path)
        break

# the array needs to be transformed in a Dataframe to save in a .CSV file
df_com = pd.DataFrame(combine, columns=["path","ethnicity","gender","age"])

df_com.to_csv('data_tot.csv', index = True)
```
